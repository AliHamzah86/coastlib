# Extreme Value Analysis Basic Tutorial
This tutorial covers a basic use case of the coastlib.EVA toolkit. The EVA is a class written in Python3.6+ to provide an engineer with simple syntax and shortcuts to perform complex extreme value analysis using various techniques. The best way to learn a new tool is to use it in a real scenario - in this case we are going to analyze water levels at the NOAA Battery station in New York. Since we are interested in statistical properties of water levels, we need to remove trends from the data. In this case this was already done - sea level rise and tidal variation were removed from the data, leaving water level residuals, which are assumed to be stationary in time (not fully true, but a good enough assumption for this test). Lets dive into the analysis.

## Load data

First, we need to import requried modules and load data. Pre-processed data is available at https://gitlab.com/georgebv/coastlib/raw/developer/coastlib/stats/test_data.csv. Let's also plot Hurricane Sandy, which hit New York on October 29 2012.

```python
import matplotlib.pyplot as plt
from coastlib.stats.extreme import EVA
import pandas as pd

# Ignore warnings since this is a tutorial. Warnings are from scipy.stats.* trying to fit MLE using invalid parameters
# (e.g. scale < 0) and its internal optimizer throwing out warnings for receiving nan's
# (optimizer is general and is not related to MLE)
import warnings
warnings.filterwarnings('ignore')

# Don't forget to tell pandas that first column is index and to parse i as datetime
data = pd.read_csv(
    r'https://dl.dropboxusercontent.com/s/xgqp2v2f1foyrvb/battery_residuals.csv?dl=0',
    index_col=0, parse_dates=True
)
```

```python
mask = (data.index >= pd.Timestamp(year=2012, month=10, day=22)) & (data.index <= pd.Timestamp(year=2012, month=11, day=5))
sandy = data[mask]
sandy.plot()
```

## Perform Extreme Value Analysis
### Initialize the EVA object
The first in every analysis is initializing the EVA object. This simple command loads data, processes it, and determines parameters which are going to be further used in the analysis. We pass data to **dataframe** and tell which column to analyze by passing the column name to **column** (if not passed, by default it takes the first column). **Block size** determines how we define return periods - it has units of days and by passing the value 365.25 we say that return period of 1 means 1-year (if we wanted monthly return periods, we would pass 365.25/12). **Gap length** has units of hours and determines at what point gaps are accounted for - e.g. if we have two chunks of data each being 3 months long, then, if they are separated by a gap of, say, 5 months, they are treated as 6 months of data (if gap_length=None, then they would be treated as 11 months of data). Effectively, EVA finds gaps longer than gap_length and eliminates them by shifting all data "right" (future) of the gap to the "left" (past). This has a large effect on estimation of probabilities and return periods, especially in data sets with gaps (such as in this case where we have severl long gaps). Without gap_length return periods would be less conservative with a weak assumption of no extreme events occuring during periods with no data.

```python
analysis = EVA(dataframe=data, column='Residuals (ft)', block_size=365.2425, gap_length=24)
```

### Extract extreme values
Now to the fun part - let's extract extreme values. For this basic case we will use Block Maxima (BM). Block maxima uses block sizes defined during initializing the EVA object - in this case its one Gregorian year. When extreme values are extracted they are also assigned a return period (probability of exceedance). This is done using a specified plotting position (by default its Weibull, [read more here](https://matplotlib.org/mpl-probscale/tutorial/closer_look_at_plot_pos.html)). In this example we pass it explicitly. See code docstrings for available plotting positions. Plotting position has no effect on inference or extreme value prediction - only on visual representation of return value plot shown later and some subjective goodness-of-fit statistics. Let's also plot the extreme values - it shows extracted extreme values and block boundarier. The plot is interactive and displays information when hovering over points (interactive plots do not work well in a Jupyter notebook - run the model from Python directly for full functionality).

```python
analysis.get_extremes(method='BM', plotting_position='Weibull')
analysis.plot_extremes()
```

### Fit the distribution
For the Block Maxima extreme value extraction method it is common practice to use the Generalied Extreme Value distribution ([read more here](https://en.wikipedia.org/wiki/Generalized_extreme_value_distribution)). The GEV distribution encapsulates three classic distributions ,which, as was theoretically proven, cover all possible distributions of block maxima extremes. The three types are Gumbel, Frechet, and Weibull. Our first guess will be very general and light on assumptions - the general 3-parameter GEV distribution. Let's fit it using Maximum Likelihood Estimate (MLE), estimate confidence intervals using the Delta method, and plot results.

```python
# Dsitribution names come form scipy.stats, see https://docs.scipy.org/doc/scipy/reference/stats.html
analysis.fit(distribution_name='genextreme', fit_method='MLE')

# Return value (summary) plot
# See available confidence interval estimation methods in EVA docstrings (there are plenty and Delta is far from best)
analysis.generate_results(alpha=.95, ci_kwargs=dict(method='Delta'))

# See available plotting positions in EVA docstrings
analysis.plot_summary(bins=10, plotting_position='Cunnane')

# QQ plot
analysis.plot_qq(quantiles=True, plotting_position='Cunnane', k=3)

# PP plot
analysis.plot_qq(quantiles=False, plotting_position='Cunnane', k=3)
```

Lets see Akaike Information Criteria (AIC, [see more here](https://en.wikipedia.org/wiki/Akaike_information_criterion)) and analysis summary to check if our model is adequate and not potentially degenerate.

```python
# k stands for number of parameters, order=2 for small samples
print(f'AIC = {analysis.goodness_of_fit(method="AIC", order=2, k=3):.2f}')
# Print the EVA object instance at any time to see a brief summary
print(analysis)
```

### Test for degeneracy
Now lets see if the model can be simplified and approximated as a 2-parameter Gumbel distribution. This is equivalent to a GEV model degenerate in its shape parameter. Let's repeat the same actions as before but with a fixed shape parameter. EVA remembers choice of fixed parameters and carries if over to other calculation steps. Read [scipy.stats documentation](https://docs.scipy.org/doc/scipy/reference/stats.html) to understand fixed parameter syntax.

```python
# Dsitribution names come form scipy.stats, see https://docs.scipy.org/doc/scipy/reference/stats.html
analysis.fit(distribution_name='genextreme', fit_method='MLE', scipy_fit_options=dict(f0=0))

# Return value (summary) plot
# Delta is not currently implemented for fixed-parameter scenarios
analysis.generate_results(alpha=.95, ci_kwargs=dict(method='Monte Carlo', k=100))

# See available plotting positions in EVA docstrings
analysis.plot_summary(bins=10, plotting_position='Cunnane')

# QQ plot
analysis.plot_qq(quantiles=True, plotting_position='Cunnane', k=2)

# PP plot
analysis.plot_qq(quantiles=False, plotting_position='Cunnane', k=2)

# k stands for number of parameters (we have fixed shape, so now its 2), order=2 for small samples
print(f'AIC = {analysis.goodness_of_fit(method="AIC", order=2, k=2):.2f}')
# Print the EVA object instance at any time to see a brief summary
print(analysis)
```

AIC suggests that the Gumbel and GEV models are equivalent based on values of penalized likelihood. The GEV model is chosen in this scenario since it better fits the observed extreme values and gives more conservative estimates without introducing excessive bias.
### Perform a production run of the model
Now, after we are confident in our model choice, its time to perform the production run of the model. The only difference from previous steps is that we perform a proper Monte Carlo simulation with a large number of samples to get an accurate estimate of confidence intervals. This is much more accurate than the Delta method shown earlier, since MC doesn't make any assumptions about the return value distribution (Delta assumes normality, which is clearly violated as shown below).

```python
analysis.fit(distribution_name='genextreme', fit_method='MLE')
analysis.generate_results(
    alpha=.95,
    ci_kwargs=dict(method='Monte Carlo', k=1e3, source='data', sampling_method='constant', assume_normality=False)
)
analysis.plot_summary(bins=10, plotting_position='Cunnane')
```

Analysis results can be exported to an excel file by calling `analysis.results.to_excel(path/to/excel.xlsx)`. For more export options see [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/api.html#id12). If results for more return periods is needed, an `rp` argument with an array of return periods can be passed to `analysis.generate_results(rp=array)`.
